```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
crop_data = pd.read_csv("Crop_recommendation.csv")

print("Dataset loaded successfully")
print(crop_data.head())

# Graph 1: Crop Distribution
crop_data['label'].value_counts().plot(kind='bar')

plt.title("Crop Distribution")
plt.xlabel("Crop Name")
plt.ylabel("Count")

plt.show()

# Graph 2: Correlation Heatmap
plt.figure(figsize=(10,8))

sns.heatmap(
    crop_data.corr(numeric_only=True),
    annot=True
)

plt.title("Feature Correlation Heatmap")

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Input and output data
features = crop_data.drop('label', axis=1)
target = crop_data['label']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.2,
    random_state=42
)

# Create and train model
model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

print("Training model...")

model.fit(X_train, y_train)

# Accuracy check
predictions = model.predict(X_test)

accuracy = accuracy_score(y_test, predictions)

print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
import joblib

# Save model
joblib.dump(model, "crop_model.pkl")

print("Model saved successfully")

# Sample prediction
sample_data = [[90, 42, 43, 20.8, 82, 6.5, 202]]

result = model.predict(sample_data)

print("Recommended Crop:", result[0])
```

```python
sample_data = pd.DataFrame(
    [[90, 42, 43, 20.8, 82, 6.5, 202]],
    columns=['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
)

result = model.predict(sample_data)

print("Recommended Crop:", result[0])
```
